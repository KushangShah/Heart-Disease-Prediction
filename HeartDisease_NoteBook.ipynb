### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

### Hypothetical Statement - 1: Association between Sex and Heart Disease

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

- **Null (H‚ÇÄ)**: There is no association between sex (male/female) and presence of heart disease.
- **Alternative (H‚ÇÅ)**: There is an association between sex and presence of heart disease

#### 2. Perform an appropriate statistical test.

```python
# Perform Statistical Test to obtain P-Value
contingency = pd.crosstab(df['sex'], df['target'])

# Perform Chi-square test
chi2, p, dof, expected = stats.chi2_contingency(contingency)
print("Chi-square Statistic:", chi2)
print("p-value:", p)

# interpretration of p value
alpha = 0.05
if p < alpha:
    print("Reject the null hypothesis: There is a significant association between sex and target.")
else:
    print("Fail to reject the null hypothesis: No significant association between sex and target.")
```

##### Which statistical test have you done to obtain P-Value?

I used Chi-square test to check weather the distribution of one categorical variable differes depending on the other independent.

##### Why did you choose the specific statistical test?

Since both variables(sex and target) are categorical, and knowing sex difference helps personalize programs and risk communication. But, Overemphasis could bias diagnosis if other factors (age, lifestyle) aren‚Äôt considered

### Hypothetical Statement - 2: Difference in Mean Cholesterol between Disease and Non-Disease

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

- **Null (H‚ÇÄ)**: Mean cholesterol level is the same for people with and without heart disease.
- **Alternative (H‚ÇÅ)**: Mean cholesterol levels differ between the two groups.

#### 2. Perform an appropriate statistical test.

```python
# Perform Statistical Test to obtain P-Value
chol_disease = df[df['target'] == 1]['cholesterol']
chol_no_disease = df[df['target'] == 0]['cholesterol']

# Perform independent t-test
t_stat, p_val = stats.ttest_ind(chol_disease, chol_no_disease, equal_var=False)
print("T-statistic:", t_stat)
print("p-value:", p_val)

# interpretation
alpha = 0.05
if p_val < alpha:
    print("Reject the null hypothesis: Mean cholesterol levels differ between the two groups.")
else:
    print("Fail to reject the null hypothesis: Mean cholesterol level is the same for people with and without heart disease.")
```

##### Which statistical test have you done to obtain P-Value?

I performed t-test to check if two group means are significantly different.

##### Why did you choose the specific statistical test?

The dependent variable(cholestrol) is continuous and the grouping variable(target) is variable.

## ***7. ML Model Implementation***

### ML Model - 1: Logistic Regression

```python
# ML Model - 1 Logistic regression implementation
log_reg = LogisticRegression(random_state=42, max_iter=1000)

# Fit the Algorithm
log_reg.fit(X_train_scaled, y_train)

# Predict on the model
y_pred_scal = log_reg.predict(X_test_scaled)
y_proba_scal = log_reg.predict_proba(X_test_scaled)[:, 1]
```

#### 1. Explain the ML Model used and it's performance using Evaluation metric Score Chart.

```python
metric_functions = {
    'accuracy': accuracy_score,
    'precision': precision_score,
    'recall': recall_score,
    'f1': f1_score,
    'roc_auc': roc_auc_score
}

# evaluation metrics
metrics = ['accuracy', 'precision', 'recall', 'f1', 'roc_auc']

scal_evaluation_results = []

for metric in metrics:
    func = metric_functions[metric]  # getting the actual function
        
    if metric != "roc_auc":
        score = func(y_test, y_pred_scal)
    else:
        score = func(y_test, y_proba_scal)
    
    scal_evaluation_results.append({
        'Metric': metric,
        'Score': score
    })
        
display(scal_evaluation_results)

# plot the graph
metric_names = [result['Metric'] for result in scal_evaluation_results]
metric_scores = [result['Score'] for result in scal_evaluation_results]

plt.figure(figsize=(10,5))
sns.barplot(
    x=metric_names,
    y=metric_scores,
    hue=metric_names,
    palette='viridis'
    )

# Display mean values on bars
for i, v in enumerate(metric_scores):
    plt.text(i, v+0.01, f'{v:.3f}', ha='center', color='white', weight='bold', fontsize=10)

plt.title("Logistic Regression Evaluation Metrics")
plt.ylim(0, 1)
plt.ylabel("Score")
plt.show()
```

#### 2. Cross- Validation & Hyperparameter Tuning

##### Cross - Validation

```python
metrics = ['accuracy', 'precision', 'recall', 'f1', 'roc_auc']

# Create an empty list to store results
log_reg_cross_val = []

# Cross-validation for each metric
log_reg = LogisticRegression(random_state=42, max_iter=1000)

for metric in metrics:
    cv_scores = cross_val_score(log_reg, X_train_scaled, y_train, cv=5, scoring=metric)
    mean_score = cv_scores.mean()
    
    # Print (optional)
    print(f"\nCross-validated {metric} scores: {cv_scores}")
    print(f"Logistic Regression - Mean CV {metric}: {mean_score:.4f}")
    
    # Store in list
    log_reg_cross_val.append({
        'Metric': metric,
        'Fold 1': cv_scores[0],
        'Fold 2': cv_scores[1],
        'Fold 3': cv_scores[2],
        'Fold 4': cv_scores[3],
        'Fold 5': cv_scores[4],
        'Mean': mean_score
    })

# Convert to DataFrame
log_reg_cv_df = pd.DataFrame(log_reg_cross_val)
display(log_reg_cv_df)
log_reg_cv_df.to_csv("metricx_insight/log_reg_cv_df.csv")
```

##### Hyperparameter Tunning

```python
# hyperparameter tunning using gridseachCV
param_grid = {
    'C': [0.01, 0.1, 1, 10, 100],
    'penalty': ['l1', 'l2'],
    'solver': ['liblinear', 'saga']
}

# Create an empty list to store results
log_reg_ht_results = []

for metric in metrics:
    grid_log_reg = GridSearchCV(
        LogisticRegression(random_state=42, max_iter=1000),
        param_grid,
        scoring=metric,
        cv=5,
        n_jobs=-1
    )
    grid_log_reg.fit(X_train_scaled, y_train)
    
    mean_score = grid_log_reg.best_score_
    
    # Extract per-fold results
    fold_scores = [grid_log_reg.cv_results_[f'split{i}_test_score'].mean() for i in range(5)]
    
    print(f"\n===== Metric: {metric.upper()} =====")
    print("Best Parameters:", grid_log_reg.best_params_)
    # print(grid_log_reg)
    print("Best CV Mean Score:", round(mean_score, 4))
    
    log_reg_ht_results.append({
        'Metric': metric,
        'Fold 1': fold_scores[0],
        'Fold 2': fold_scores[1],
        'Fold 3': fold_scores[2],
        'Fold 4': fold_scores[3],
        'Fold 5': fold_scores[4],
        'Mean': mean_score
    })

log_reg_hyper_tunning_df = pd.DataFrame(log_reg_ht_results)
display(log_reg_hyper_tunning_df)
log_reg_hyper_tunning_df.to_csv("metricx_insight/log_reg_hyper_tunning_df.csv")
```

Then, it simply predicts that the new patient has heart disease if most of those 'k' similar patients had it, and predicts they don't if most of those neighbors didn't. It's like saying, "Show me your friends, and I'll tell you who you are!"<br>
Now, for which features were most important, KNN doesn't exactly tell us directly like some other models do. So, even though KNN works differently, we can feel pretty confident that these same features are likely the ones it's relying on heavily to make its predictions too. If we needed a super deep dive into why KNN made a specific prediction for a single patient, we could use some special tools called SHAP or LIME, but based on our overall analysis, those features I mentioned are the big players.

### 3. Explain the model which you have used and the feature importance using any model explainability tool?

When we think about how this project can actually help people, especially in healthcare, we need to look at a few key numbers:

*   **Recall:** This is super important! It tells us how good our model is at catching *all* the actual heart disease cases. We really don't want to miss anyone who might be at risk, right?
*   **Precision:** This one is about not raising false alarms. While catching everyone is the priority, we also don't want to tell someone they might have heart disease when they don't, as that can cause a lot of worry and lead to unnecessary tests.
*   **ROC-AUC:** Think of this as an overall score of how well our model can tell the difference between someone with heart disease and someone without. A higher number here means it's doing a pretty good job at separating the two groups.
*   **Accuracy:** This just tells us how often our model is right overall. It's a good general measure, but for something serious like heart disease, we need to look deeper at recall and precision too.
